## 2. Outliers checking/treatment

```python
#detect Outliers 
df_housing.plot(kind='box' , subplots = True , figsize = (20,20) , layout = (3,2))
plt.show()
```

```python
#treatment of outliers
num_col = ['price','area','bedrooms','bathrooms','stories','parking']
def plot_num(num):
    plt.figure(figsize=(6,6))
    plt.title(num)
    sns.boxplot(df_housing[num])
    plt.show()
    return
```

```python
for num in num_col:
    q3,q1 = np.percentile(df_housing[num],[75,25])
    q4 = np.percentile(df_housing[num],[95])
    df_housing.loc[df_housing[num] > q4[0] , num] = q4[0]
    plot_num(num)
```

```python
df_housing.head(10)
```

```python
#create dummy values
dummies = pd.get_dummies(df_housing['furnishingstatus'])
dummies
```

```python
df_housing = pd.concat([df_housing,dummies],axis=1)
df_housing
```

```python
#drop column
df_housing = df_housing.drop(['furnishingstatus'],axis=1)
df_housing
```

#### Observations :
##### - here no missing values in data  
##### - Detect the outliers
##### - Treatement of outliers using capping method
##### - Here one column is furnishingstatus and this colomn is not encoded by 0 and 1 .so we create dummies values of column.
##### - furnishingstatus column is not important for our data so we drop the column

# 4.Data Preparation

## Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
columns = ['mainroad','guestroom','basement','hotwaterheating',
          'airconditioning','prefarea']
df_housing[columns] = df_housing[columns].apply(LabelEncoder().fit_transform)
df_housing.head(10)
```

## train_test_split

```python
X = df_housing.drop('price',axis=1)
Y = df_housing['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
x_train
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
std_scale = MinMaxScaler()
```

```python
x_train[[ 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']] = std_scale.fit_transform(x_train[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']])
x_train.head()
```

```python
x_train = std_scale.fit_transform(x_train)
x_test = std_scale.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
#r2 score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
